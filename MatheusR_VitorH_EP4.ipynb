```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Data.csv')
```

```python
candidate = {}
President = df['NM_VOTAVEL']
```

```python
for x in President:
    if x not in candidate:
        candidate[x] = 0
candidate
```

```python
zone = {}
dataZone = df["NR_ZONA"]
for x in dataZone:
    if x not in zone: zone[x] = {'Branco': 0, 'Nulo': 0, 'CIRO GOMES': 0, 'FERNANDO HADDAD': 0, 'HENRIQUE MEIRELLES': 0, 'VERA': 0, 'JAIR BOLSONARO': 0, 'MARINA SILVA': 0, 'ALVARO DIAS': 0, 'JOAO AMOEDO': 0, 'GERALDO ALCKMIN': 0, 'GUILHERME BOULOS': 0, 'CABO DACIOLO': 0, 'EYMAEL': 0, 'JOAO GOULART FILHO': 0}
zone
```

```python
lengthRows = len(df["NM_VOTAVEL"])
for row in range(0, lengthRows):
    zone[df["NR_ZONA"][row]][df["NM_VOTAVEL"][row]] += df["QT_VOTOS"][row]
zone
```

```python
sorted1 = sorted(zone[412].items(), key=lambda kv: kv[1])
sorted2 = sorted1[-5::]
keys = []
values = []
for x in sorted2:
    keys.append(x[0])
    values.append(x[1])
print(pd.DataFrame(sorted1, columns=["Candidatos", "Votos"]))
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,2,2])
ax.bar(keys,values)
ax.set_title("412 Zone", fontsize=(10))
ax.set_ylabel("Vote", fontsize=(10))
ax.set_xlabel("candidate", fontsize=(10))
plt.show()
```
